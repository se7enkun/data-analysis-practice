```python
import pandas as pd
data1 = pd.read_excel('商品销售数据.xlsx', sheet_name='信息表')
data1
```

```python
data2 = pd.read_excel('商品销售数据.xlsx', sheet_name='销售数据表')
data2
```

```python
data_all = pd.merge(data1, data2, on='商品编号', how='right')
data_all
```

```python
data_all['销售金额'] = data_all['订单数量'] * data_all['商品销售价']
data_all
```

```python
data_all['月份'] = data_all['订单日期'].dt.month
data_all
```

```python
data_all_1 = data_all.groupby(by='月份', as_index=False)['销售金额'].sum()
data_all_1
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 6))
x = data_all_1['月份']
y = data_all_1['销售金额']
plt.plot(x, y, color='k', linewidth=3, linestyle='solid', marker='s', markersize=8)
plt.title(label='商品每月销售金额折线图', font='KaiTi', fontsize=30, color='k', loc='center')
plt.xlabel('月份', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.ylabel('销售金额(元)', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.xticks(x, ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月'])
plt.ylim(300000, 600000)
plt.show()
```

```python
data_all_2 = data_all.groupby(by=['月份', '商品大类'], as_index=False)['销售金额'].sum()
list1 = data_all_2[data_all_2['商品大类'] == '运动上装']
list2 = data_all_2[data_all_2['商品大类'] == '运动下装']
list3 = data_all_2[data_all_2['商品大类'] == '运动配饰']
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 6))
x = list1['月份']
y1 = list1['销售金额']
y2 = list2['销售金额']
y3 = list3['销售金额']
plt.bar(x - 0.2, y1, width=0.2, color='k', label='运动上装')
plt.bar(x, y2, width=0.2, color='b', label='运动下装')
plt.bar(x + 0.2, y3, width=0.2, color='r', label='运动配饰')
plt.title(label='运动上装、运动下装和运动配饰月度销售金额对比图', font='KaiTi', fontsize=30, color='k', loc='center')
plt.xlabel('月份', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.ylabel('销售金额(元)', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.legend(loc='upper right', fontsize=12)
plt.xticks(x, ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月'])
plt.ylim(0, 400000)
plt.show()
```

```python
data_all_3 = data_all.groupby(by='商品小类', as_index=False)['销售金额'].sum()
data_all_3
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 6))
x = data_all_3['商品小类']
y = data_all_3['销售金额'] / 10000
plt.bar(x, y, width=0.6, color='k')
plt.title(label='各商品小类年度销售金额对比图', font='KaiTi', fontsize=30, color='k', loc='center')
plt.xlabel('商品小类', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.ylabel('销售金额(万元)', font='KaiTi', fontsize=15, color='k', labelpad=5)
plt.ylim(0, 120)
for i, j in zip(x, y):
    plt.text(i, j, f'{j:.2f}', ha='center', va='bottom', size=15)
plt.show()
```
